```python
import pandas as pd, seaborn as sns

df = pd.read_excel("DailyActivities.xlsx").set_index('Area of Interest')

plot = sns.heatmap(
    df,
    cmap="flare",
    annot=True,
    vmin=0, vmax=10)

plot.set(xlabel="People in the Dataset")

plt.show()
```

```python
import pandas as pd, seaborn as sns

df = pd.read_excel("DailyActivities.xlsx")

df = pd.melt(
    df,
    id_vars='Area of Interest',
    var_name='Person',
    value_name='Time')

df = df[[
    'Area of Interest',
    'Person',
    'Time']]

plot = sns.barplot(
    x= "Person",
    y = "Time",
    hue = "Area of Interest",
    data = df)

plot.set(
    xlabel="People in the Dataset",
    ylabel="Time Had Been Spent")

plt.show()
```
